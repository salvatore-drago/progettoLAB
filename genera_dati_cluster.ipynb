```python
import numpy as np
from pyod.utils.data import generate_data_clusters
x, y= generate_data_clusters(n_train=2000, n_test=50, n_clusters=2, n_features=3, contamination=0.1, size='different', density='different', dist=0.80, random_state=1, return_in_clusters=True)
Intrusi=[]
Cluster_1=[]
Cluster_2=[]
```

```python
intrusi_1=0
intrusi_2=0
y1_list=y[0].tolist()
y2_list=y[1].tolist()
for i in range(0, len(y1_list)):
    if y1_list[i]==1.0:
        intrusi_1=intrusi_1+1

for i in range(0, len(y2_list)):
    if y2_list[i]==1.0:
        intrusi_2=intrusi_2+1

print(intrusi_1)
print(intrusi_2)
```

```python
x_list=x[0].tolist()
n=len(x_list)
for i in range(n-intrusi_1,n):
    Intrusi.append(x_list[i])

Cluster_1=x_list[0:n-intrusi_1]
print(Intrusi)
#print(Cluster_1)
```

```python
z_list=x[1].tolist()
n=len(z_list)
for i in range(n-intrusi_2,n):
    Intrusi.append(z_list[i])

Cluster_2=z_list[0:n-intrusi_2]
print(Intrusi)
#print(Cluster_2)

```

```python
print(len(Intrusi))
print(len(Cluster_1))
print(len(Cluster_2))
```

```python
for i in range(0, len(Intrusi)):
    a=Intrusi[i]
    a.append(1)
    Intrusi[i]=a
for i in range(0, len(Cluster_1)):
    b=Cluster_1[i]
    b.append(0)
    Cluster_1[i]=b
for i in range(0, len(Cluster_2)):
    c=Cluster_2[i]
    c.append(0)
    Cluster_2[i]=c


    
```

```python
import random
random.shuffle(Intrusi)
#print(Intrusi)
c1=(Cluster_1+Intrusi[0:int(len(Intrusi)/2)])
c2=(Cluster_2+Intrusi[int(len(Intrusi)/2):len(Intrusi)])
random.shuffle(c1)
random.shuffle(c2)
print(len(c1))
print(len(c2))
```

```python
# creazione dataset con sudden e recourring concept drift
import pandas as pd
Dataset=c1[0:int(len(c1)/2)]+c2+c1[int(len(c1)/2)+1:len(c1)]
print(len(Dataset))
Dataset_np=np.array(Dataset)
pd.DataFrame(Dataset).to_csv('SRdrift.csv')
```
